```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the dataset into a pandas dataframe
df = pd.read_csv(r"C:\Users\RAHUL ADITYA\Documents\ICT\titanic_dataset.csv")

# Drop irrelevant columns from the dataframe
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# Replace missing values in 'Age' and 'Embarked' columns with their respective mode values
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Convert categorical variables into numerical variables using LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])

# Scale the continuous variables using StandardScaler
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

# Split the dataset into input features (X) and target variable 👍
X = df.drop('Survived', axis=1)
y = df['Survived']

# Define the kNN model
knn_model = KNeighborsClassifier(n_neighbors=5)

# Define the SVM model
svm_model = SVC(kernel='linear', C=1)

# Define the number of folds for k-fold and stratified k-fold cross-validation
num_folds = 10

# Define the k-fold cross-validation technique
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Define the stratified k-fold cross-validation technique
stratified_kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

# Define a function to calculate the average accuracy score of a model using a given cross-validation technique
def cross_val_score(model, X, y, cv):
    scores = []
    for train_index, test_index in cv.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        scores.append(accuracy_score(y_test, y_pred))
    return np.mean(scores)

# Calculate the average accuracy scores of kNN and SVM models using k-fold and stratified k-fold cross-validation techniques
knn_kfold_score = cross_val_score(knn_model, X, y, kfold)
knn_stratified_kfold_score = cross_val_score(knn_model, X, y, stratified_kfold)
svm_kfold_score = cross_val_score(svm_model, X, y, kfold)
svm_stratified_kfold_score = cross_val_score(svm_model, X, y, stratified_kfold)

# Print the average accuracy scores of kNN and SVM models using k-fold and stratified k-fold cross-validation techniques
print("kNN model accuracy using k-fold cross-validation: ", knn_kfold_score)
print("kNN model accuracy using stratified k-fold cross-validation: ", knn_stratified_kfold_score)
print("SVM model accuracy using k-fold cross-validation: ", svm_kfold_score)
print("SVM model accuracy using stratified k-fold cross-validation:",svm_stratified_kfold_score)
```
